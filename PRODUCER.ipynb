```python
#PRODUCER
from confluent_kafka import Producer
producer = Producer(**config)

```

```python
# Delivery report callback for produced messages
import json
def delivery_report(err, msg):
    """ Called once for each message produced to indicate delivery result.
        Triggered by poll() or flush(). """
    if err is not None:
        print('Message delivery failed: {}'.format(err))
    else:
        print('Message delivered to {} [{}]'.format(msg.topic(), msg.partition()))

# Produce messages to the 'users' topic
topic = 'topic_0'

try:
    for timestamp, values in data['Time Series (5min)'].items():
        record_key = timestamp
        record_value = json.dumps(values)
        producer.produce(topic, key=record_key, value=record_value, callback=delivery_report)
        # Wait up to 1 second for events. Callbacks will be invoked during this method call if the message is acknowledged.
        producer.poll(1)
except Exception as e:
    print(f'Failed to produce message: {e}')

# Wait for any outstanding messages to be delivered and delivery report callbacks to be triggered.
producer.flush()
```
